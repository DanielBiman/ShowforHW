### Exercise 11.1

Write a program that reads the contents of the file "`blakepoems.txt`", splits it into words (where everything that is not a letter is considered a word boundary), and case-insensitively builds a dictionary that stores for every word how often it occurs in the text. Then print all the words with their quantities in alphabetical order.

```python
# Counting words in blakepoems.txt.
#split into words
#only letters
#build dictionary that counts every word and prints this in alphabetical order

text = open("blakepoems.txt")
text = text.read()

#Sorting
text = text.split()
text = sorted([char.lower() for char in text if char.isalpha()])
#print(text)
myl = sorted(set(text))
#print(myl)
#Creating dict as result
result = dict()
for key in myl:
    result[key] = text.count(key)

#Uncomment me and you see the result:    
#result

```
